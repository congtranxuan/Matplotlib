```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Import data and merge together

```python
citypath = "Pyber_data/city_data.csv"
ridepath = "Pyber_data/ride_data.csv"
cityfile = pd.read_csv(citypath)
ridefile = pd.read_csv(ridepath)
datain = pd.merge(cityfile,ridefile, on = "city", how = "outer")
datain.head()
```

```python
GroupedTC = datain.groupby(["type","city"])
Avg_fare = GroupedTC["fare"].mean()
No_ride = GroupedTC["ride_id"].count()
No_driver_city = GroupedTC[["type","city","driver_count"]].head(1).set_index(["type","city"])
No_driver = No_driver_city.sort_values(["type","city"],ascending = True)*16
```

```python
fig,ax = plt.subplots(figsize = (8,6))
facecolor = ["gold","skyblue","coral"]
typ = No_driver.index.levels[0]
typ.columns=["City types"]
for i in range(len(typ)):
    plt.scatter(No_ride.loc[typ[i]],Avg_fare.loc[typ[i]],No_driver.loc[typ[i]],label = typ[i],marker ="o",facecolor =facecolor[i],edgecolor="k",alpha = 0.8,linewidth =1)
ax.legend(loc="best", fontsize="small", fancybox=True)
ax.set_xlabel("Total Number of Rides (Per City)")
ax.set_ylabel("Average Fare ($)")
ax.set_title("Pyber Ride Sharing Data (2016)")
plt.grid()
plt.tight_layout()
plt.savefig("Images/Pyberdata2016.jpg")
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure
```

```python
Totalfare = datain.groupby(["type"])["fare"].sum()
Totalfare
```

```python
fig,ax = plt.subplots(figsize =(6,6))
color = ["Gold","Skyblue","Coral"]
explode = [0,0,0.1]
labels = Totalfare.index
Totalfare.plot(kind="pie",colors = color,explode = explode,labels = labels,autopct="%1.1f%%", shadow = True,startangle = 150)
ax.set_ylabel("")
ax.set_title("% of Total Fares by City Type")
ax.axis("off")

plt.tight_layout()
plt.savefig("Images/Percentfares.jpg")
plt.show()
```
